# Projet Data

Nous avons un dataset d'anime que nous voulons nettoyer pour ensuite determiner quels animes mettre en avant pour un site de streaming.

Tout d'abord commencons par charger les outils dont nous auront besoin durant cette analyse :

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Maintenant nous allons charger le fichier et afficher les premières lignes : 

```python
df = pd.read_csv("animes.csv")

print("fichier chargé avec succès")
print(f"Nombre d'animes : {len(df)}")
df.head()
```

Récupérons quelques informations supplémentaires sur notre tableau : 

```python
lignes, colonnes = df.shape
print(f"Notre tableau contient : {lignes} lignes et {colonnes} colonnes.")
```

```python
print("Colonnes disponibles : ")
for i, col in enumerate(df.columns, 1): 
    print(f" {i}. {col}")
```

```python
df.describe().round(2)
```

### Nettoyage des données : 

tout d'abord on va extraire le genre principal des animes :

```python
df['Genre_Principal'] = df['Genre_Tags'].str.split(' / ').str[0]
print(df[['Anime', 'Genre_Principal']].head(10))
```

maintenant on va filtrer les anime avec une note supérieure à 8.5 et qui sont fini, en effet ce serait les plus intéressants pour un site de streaming.

```python
condition1 = df['Note_Globale'] > 8
condition2 = df['Status'] == 'Fini'
excellents = df[condition1 & condition2]

print(f"{len(excellents)} animes ont une note supérieure à 8.5")
print(excellents[['Anime', 'Note_Globale']].to_string(index=False))
```

Voyons voir les anime les mieux notés : 

```python
note_max = df['Note_Globale'].max()

meilleur = df[df['Note_Globale'] == note_max]['Anime']

print(meilleur.to_string(index=False))
```

### Visualisation :

Voyons voir un histogramme des notes : 

```python
plt.figure(figsize=(10,5))

plt.hist(df['Note_Globale'], bins=10, color='#ff0000ff', edgecolor='black')

moyenne = df['Note_Globale'].mean()
plt.axvline(moyenne, color='green', linestyle='--', linewidth=2, label=f'Moyenne : {moyenne: .1f}')

plt.title("Histogramme des notes d'anime", fontsize=14, fontweight='bold')
plt.xlabel("Note")
plt.ylabel("Nombres d'animes")
plt.legend()
plt.show()
```

Voyons voir un graphique qui va nous aider a voir la repartition des notes en fonction du nombre d'épisodes :

```python
plt.figure(figsize=(10, 6))

df_filtre = df[df['Nb_Episodes'] <= 500]

plt.scatter(df_filtre['Nb_Episodes'], df_filtre['Note_Globale'], alpha=0.4, c='magenta', s=80, edgecolor='white')

plt.title("Note vs Nombres d'épisodes", fontsize=14, fontweight='bold')
plt.xlabel("Nombre d'épisodes")
plt.ylabel("Note Global")
plt.grid(True, alpha=0.3)
plt.show()
```
